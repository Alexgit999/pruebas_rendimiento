```python
import pandas as pd
import pyodbc
from sqlalchemy import create_engine
```

```python
##  Cargamos y transformamos los datos en un DataFrame  ##
df_fichero = pd.read_csv('C:\\Users\\amerino.practicas\\Documents\\python\\export_csv\\fichero.csv', sep= '|', on_bad_lines='skip')

##  Limpieza de datos  ##
# Quitamos columnas "Unnamed"
df_fichero = df_fichero.loc[:, ~df_fichero.columns.str.contains('^Unnamed')].reset_index(drop=True)
# Detectar columnas duplicadas (en éste caso no hay, si las hubiere, tendríamos que ir fusionando los datos en la primera columna no duplicada de la lista)
#col_duplicated = df_fichero.loc[:,~df_fichero.columns.duplicated()].copy() # only first non duplicated name columns
# 
col_duplicated = df_fichero.loc[~df_fichero.index.duplicated(),:].copy()

col_duplicated.drop(col_duplicated.index[8393], axis=0, inplace=True) # último registro borrado, era nulo
print(df_fichero)
col_duplicated
```

```python
##  Conexión con la base de datos  ##
# Volcar los datos a viajes_testing

server = "Unificada"
database = "ASIGNA"
 
conn = pyodbc.connect("DRIVER={SQL Server};SERVER=" + server + ";trusted_connection=yes;")
#podría no ser necesario (funcionan independientemente), aunque lo suyo es usar conn y create_engine por si acaso

# connection_string
engine = create_engine(f"mssql+pyodbc://@{server}/{database}?trusted_connection=yes&driver=SQL+Server")
engine = engine.connect()

# Nos aseguramos que estamos conectados con ASIGNA y vemos las tablas que contiene
df_schemas = pd.read_sql('SELECT * FROM INFORMATION_SCHEMA.TABLES;', con=engine)
df_schemas
```
