```python
# Función para convertir columnas automáticamente
def convertir_columnas(df_fichero):
    for col in df_fichero.columns:
        try:
            # Intentar convertir a int
            df_fichero[col] = pd.to_numeric(df_fichero[col], errors='coerce')
            if df_fichero[col].notnull().all():  # Si todos los valores son numéricos
                df_fichero[col] = df_fichero[col].astype(int)
            else:
                # Si no todos son numéricos, mantener como texto
                df_fichero[col] = df_fichero[col].astype(str)
        except Exception as e:
            # Si falla, forzar a texto
            df_fichero[col] = df_fichero[col].astype(str)
    return df_fichero

# Convertir las columnas
df_convertido = convertir_columnas(df_fichero)
#df_convertido[df_convertido.select_dtypes(include=['object']).columns] = df_convertido.select_dtypes(include=['object']).astype(str)

df_convertido.info()
```
